```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

# Canny

```python
img_color = cv2.imread('emma_stone.jpg', cv2.IMREAD_COLOR)
plt.figure
plt.imshow(img_color) # cmap으로 color 살릴수 없나?
```

cv2.Canny(img, threshold1, threshold2, edges=None, apertureSize=None, L2gradient=None)  
- 입력 img   
- 최소, 최대 threshold  
- edge에 Canny 결과를 저장할 변수를 적어준다.(변수로 리턴받기가 더 쉽다)  
- apertureSize 이미지 gradient를 구할 때 사용하는 sobel kernel 크기 default=3
- L2  

```python
img_canny = cv2.Canny(img_color, 50, 150)
plt.figure
plt.imshow(img_canny) 
```

팔자주름을 잘 잡도록 threshold값을 바꿔서 해보자

```python
step = 20
src = 30
dst = 150
img_canny_list = []
for i in range(src,dst,step):
    img_canny_list.append(cv2.Canny(img_color,i,dst))
    
n = 5
plt.figure(figsize= (20,20))
for i in range(n):
    ax = plt.subplot(2, n, i + 1)
    plt.imshow(img_canny_list[i])
    plt.title("original")

plt.show()
```
