```python
import matplotlib.pyplot as plt
import numpy as np

hsv = plt.get_cmap("hsv") # 이건 뭐지??
```

```python
fig, ax = plt.subplots(figsize=(5, 5))

num = 36
d_theta = 2 * np.pi / num

for i in range(num):
    x = np.sin(d_theta * i)
    y = np.cos(d_theta * i)
    ax.scatter(x, y, s = 200, fc=hsv(i/num), ec="k") #scaater 함수에 대해서 
```
