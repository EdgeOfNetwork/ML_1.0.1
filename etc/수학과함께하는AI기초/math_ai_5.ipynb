```python
import pandas as pd

df_test = pd.read_csv('fashion-mnist_test.csv')
df_train = pd.read_csv('fashion-mnist_train.csv')
print(df_train.info(),'\n')
print(df_test.info(),'\n')
print(df_train.shape,'\n')
print(df_test.shape,'\n')
```

```python
""" 각 데이터 프레임의 값을 배열에 저장 """
import numpy as np
data_train = np.array(df_train, dtype = np.float32) #csv를 배열로 저장
x_train = data_train[:, 1:]
y_train = data_train[:, 0]

data_test = np.array(df_test)
x_test = data_test[:, 1:]
y_test = data_test[:, 0]

print(df_test)
print(data_test)
```

```python
import matplotlib.pyplot as plt
#의료 종류 레이블을 딕셔너리로 저장
label_dictionary = {0:'T-shirt/top', 1:'Trouser', 2:'Pullover', 3:'Dress', 4:'Coat', 5:'Sandal', 6:'Shirt', 7:'Sneaker', 8:'Bag', 9:'Ankle boot'}

i = 234 #임의의 숫자 지정 : 6만개 데이터 중 i번째, 즉 6만개 중 내가 보고싶은 것을 아래 출력해 볼 수 있다.
plt.imshow(x_train[i].reshape(28,28), cmap = 'gray')
plt.colorbar()
plt.title('Label {},{}'.format(y_train[i], label_dictionary[y_train[i]]))
plt.show()
```

```python
for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.imshow(x_train[i].reshape(28,28))
    plt.colorbar()
    plt.title('Label {},{}'.format(y_train[i], label_dictionary[y_train[i]]))
    
plt.tight_layout()
plt.show()
```
