```python
import numpy as np
import matplotlib.pyplot as plt

#좌표 (4,3) 지정
a = np.array([4,3])

#Figure 객체를 만든 후
fig = plt.figure()

#1x1 개의 서브플롯을 추가하고, 첫번째 서브플롯을 사용한다.
ax = fig.add_subplot(1,1,1)

#변수 a의 값을 좌표로 사용하여 점을 그린다.
#a[0] 이 x 좌표, a[1]이 y좌표 s는 점의 크기
ax.scatter(a[0], a[1], s= 30)


#점 옆에 문자 a 표시, size는 문자의 크기
ax.text(a[0] + 0.2, a[1]+0.2, 'a', size=15)

#x축, y축 눈금 범위를 -5~ 5로 지정
ax.set_xticks(range(-5,6))
ax.set_yticks(range(-5,6))

#그리드 좌표축 아래에 표시
ax.grid()
ax.set_axisbelow(True)

#x축과 y축 눈금의 비율을 1:!로
#창 크기를 변경해도 비율 유지
ax.set_aspect('equal',adjustable='box')

#원점에 좌표축이 보이게
#왼쪽 y축과 아래 x축을 0,0으로 이동
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

#위쪽과 오른쪽 축을 제거
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

#그리기
plt.show()
```

```python
a = np.array([4,3])

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

#시작 위치는(0,0)이고 화살표가 그려지는 끝 위치는 변수 a의 값을 좌표로 사용하는 화살표를 그립니다.
#a[0] 과 a[1] 이 각각 끝 위치의 x좌표, y좌표입니다.
ax.quiver(0,0,a[0],a[1], angles = 'xy', scale_units='xy', scale=1)
ax.text(a[0], a[1],'a',size =15)

#x축 , y축 눈금 범위를 0에서 6으로 지정합니다.
ax.set_xticks(range(0,7))
ax.set_yticks(range(0,7))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal',adjustable = 'box')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.show()
```

```python
A = np.array([1,-2])
B = np.array([4,3])

#1개의 행과 2개의 열을 가지는 서브플롯 생성
f, ax = plt.subplots(1,2)

#인덱스를 사용하여 각 서브플롯을 사용
ax[0].title.set_text('A')
ax[1].title.set_text('B')

#각 서브플롯에 시작점이 0,0 벡터를 그림
ax[0].quiver(0,0,A[0], A[1], angles='xy', scale_units='xy', scale=1)
ax[1].quiver(0,0,B[0], B[1], angles='xy', scale_units='xy', scale=1)

#서브플롯별로 그리드를 생성
start_x = -1
end_x = 5
start_y = -3
end_y = 4
for i in range(2):
    ax[i].axis([start_x, end_x, start_y, end_y])
    ax[i].set_xticks(range(start_x, end_x))
    ax[i].set_yticks(range(start_x, end_x))
    ax[i].grid(True)
    ax[i].set_aspect('equal', adjustable='box')
    
plt.show()
```

```python
#3차원 그래프 그리기
from mpl_toolkits.mplot3d import Axes3D

a = np.array([4,5,6])

fig = plt.figure()

#subplot에서 3차원 좌표계 사용
ax = fig.add_subplot(1,1,1,projection='3d')

#시작점이 000이고 끝점좌표가 변수 a의 값인 화살표를 그림
ax.quiver(0,0,0,a[0],a[1],a[2],color='black',arrow_length_ratio=0.1)
ax.text(a[0], a[1], a[2], 'a', size= 15)

#x,y,z 좌표축 범위가 각각 0에서 8인 그리드 생성
ax.set_xlim(0,8)
ax.set_ylim(0,8)
ax.set_zlim(0,8)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.view_init(elev=20., azim=5)
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal',adjustable='box')
#어 왜 에러가뜨지? 하지만 결과는 잘 나오니까

plt.show()
```

```python
a = np.array([4,3])

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.quiver(0,0,a[0],a[1], angles = 'xy', scale_units='xy', scale=1)
#0,0에서 시작하는 선

ax.quiver(0,-3,a[0],a[1], angles = 'xy', scale_units='xy', scale=1)

ax.quiver(1,2,a[0],a[1], angles = 'xy', scale_units='xy', scale=1)

ax.quiver(-3,1,a[0],a[1], angles = 'xy', scale_units='xy', scale=1)

start = -5
end = 10

ax.set_xticks(range(start,end))
ax.set_yticks(range(start,end))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal',adjustable = 'box')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.show()
```

```python
i = np.array([1,0])
j = np.array([0,1])

#단위 벡터로 벡터 [4,3]을 나타낼 수 있다.
a = 4*i + 3*j

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.quiver(0,0,a[0], a[1], angles='xy', scale_units= 'xy', scale=1)
ax.text(a[0], a[1], '4i+3j',size= 15)

ax.set_xticks(range(0,7))
ax.set_yticks(range(0,7))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal',adjustable = 'box')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.show()
```

```python
#기본 연산 _ 덧셈
a = np.array([4,1])
b = np.array([-2,3])
sum = a+b

fig = plt.figure()
ax = fig.add_subplot(1,1,1)


ax.quiver(0,0,a[0],a[1], angles = 'xy', scale_units='xy', scale=1)
ax.text(a[0], a[1],'a',size =15)

#벡터 b를 벡터 a의 끝점에서 시작하도록 그린다.
ax.quiver(a[0],a[1],b[0],b[1], angles = 'xy', scale_units='xy', scale=1)
ax.text(a[0]+b[0], a[1]+b[1],'b',size =15)

#벡터의 합을 그리면 벡터 a의 시작점과 벡터 b의 끝점을 잇는 벡터가 그려진다.
ax.quiver(0,0 ,sum[0],sum[1], angles = 'xy', scale_units='xy', scale=1, color='blue')
ax.text(sum[0]*0.5 + 0.3,sum[1]*0.5, "a+b",size =15, color='blue')

#x축 , y축 눈금 범위를 0에서 6으로 지정합니다.
ax.set_xticks(range(0,7))
ax.set_yticks(range(0,7))
ax.grid()
ax.set_axisbelow(True)
ax.set_aspect('equal',adjustable = 'box')

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

plt.show()
```
